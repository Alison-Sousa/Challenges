```python
import pandas as pd
from google.colab import files

# Pedir ao usuário para fazer o upload do arquivo CSV
uploaded = files.upload()

# Obter o nome do arquivo CSV enviado
file_name = next(iter(uploaded))

# Ler o arquivo CSV
df = pd.read_csv(file_name)

# Exibir as 10 primeiras linhas do DataFrame com formatação ajustada
styled_df = df.head(10).style.set_properties(**{'text-align': 'left'})
styled_df
```

```python
# df é o DataFrame original e results é o DataFrame com os resultados do agrupamento
results = df.groupby(['Broker', 'Produto', 'Compra/Venda']).agg(
    Sum_Qty=('Qty', 'sum'),
    Total_Volume=('Price', lambda x: (x * df.loc[x.index, 'Qty']).sum()),
    Weighted_Avg_Price=('Price', lambda x: (x * df.loc[x.index, 'Qty']).sum() / df.loc[x.index, 'Qty'].sum())
).reset_index()

# Filtrar as linhas onde algum dos resultados é igual a 0
zero_results = results.query('Sum_Qty == 0 or Total_Volume == 0 or Weighted_Avg_Price == 0')

# Mostrar as linhas com resultados iguais a 0
print(zero_results)

# Exportar para Excel sem modificar a formatação dos números
results.to_excel('results.xlsx', index=False, engine='openpyxl')

# Baixar o arquivo Excel
from google.colab import files
files.download('results.xlsx')
```
