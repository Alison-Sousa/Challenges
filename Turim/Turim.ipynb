```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Definindo o número de pontos simulados
num_points = 100

# Simulando dados
np.random.seed(42)  # Para reprodutibilidade
usd_jpy_simulated = np.random.normal(loc=110, scale=10, size=num_points)  # Simulando USD/JPY
semiconductor_prices_simulated = np.random.normal(loc=200, scale=50, size=num_points)  # Simulando preços de semicondutores em USD
ebit_simulated = (0.5 * usd_jpy_simulated + 0.3 * semiconductor_prices_simulated + np.random.normal(loc=0, scale=20, size=num_points))  # EBIT com ruído

# Criando um DataFrame com os dados simulados
data_simulated = pd.DataFrame({
    'EBIT': ebit_simulated,
    'Semiconductor Prices (USD)': semiconductor_prices_simulated,
    'USD/JPY': usd_jpy_simulated
})

# Criar a matriz de correlação
correlation_matrix = data_simulated.corr()
print("\nMatriz de Correlação:")
print(correlation_matrix)

# Função para criar gráfico separado
def create_separate_plot(x, y, xlabel, ylabel, title):
    plt.figure(figsize=(10, 6), facecolor='none')  # Fundo da figura transparente
    sns.regplot(x=x, y=y, marker='o', color='b')
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.gca().grid(False)  # Remove a grade
    plt.gca().set_facecolor('none')  # Fundo dos eixos transparente
    for spine in plt.gca().spines.values():
        spine.set_visible(False)  # Remove todas as bordas
    plt.title(title)
    plt.savefig('plot.png', transparent=True)  # Salva o gráfico com fundo transparente
    plt.show()

# Criar gráficos separados
create_separate_plot(data_simulated['USD/JPY'], data_simulated['EBIT'], 'USD/JPY', 'EBIT', 'EBIT vs Taxa de Câmbio USD/JPY')
create_separate_plot(data_simulated['Semiconductor Prices (USD)'], data_simulated['EBIT'], 'Preços dos Semicondutores (USD)', 'EBIT', 'EBIT vs Preços dos Semicondutores (USD)')

# Gráfico 3: EBIT vs Preços dos Semicondutores (JPY)
semiconductor_prices_jpy_simulated = data_simulated['Semiconductor Prices (USD)'] * data_simulated['USD/JPY']  # Preço em JPY
create_separate_plot(semiconductor_prices_jpy_simulated, data_simulated['EBIT'], 'Preços dos Semicondutores (JPY)', 'EBIT', 'EBIT vs Preços dos Semicondutores (JPY)')

# Gráfico 4: Correlação Total
plt.figure(figsize=(10, 6), facecolor='none')  # Fundo da figura transparente
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=False)
plt.title('Matriz de Correlação')
plt.gca().set_facecolor('none')  # Fundo dos eixos transparente
for spine in plt.gca().spines.values():
    spine.set_visible(False)  # Remove todas as bordas
plt.savefig('correlation_matrix.png', transparent=True)  # Salva a matriz de correlação com fundo transparente
plt.show()
```

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import requests

# Função para obter dados da API do Banco Mundial
def get_world_bank_data(indicator, country='JP'):
    url = f"https://api.worldbank.org/v2/country/{country}/indicator/{indicator}?format=json&date=2015:2024"
    response = requests.get(url)
    data = response.json()

    if len(data) > 1:
        df = pd.DataFrame(data[1])
        df['date'] = pd.to_datetime(df['date'])
        df.set_index('date', inplace=True)
        df.rename(columns={'value': indicator}, inplace=True)
        return df[[indicator]]
    else:
        print(f"Nenhum dado encontrado para o indicador {indicator}.")
        return pd.DataFrame()

# Função para ler os dados do PMI do Excel
def get_pmi_from_excel(file_path):
    pmi_df = pd.read_excel(file_path, header=None, names=['Date', 'PMI'])

    # Limpar a coluna de data
    pmi_df['Date'] = pd.to_datetime(pmi_df['Date'].str.split('(').str[0].str.strip())
    pmi_df.set_index('Date', inplace=True)
    pmi_df['PMI'] = pd.to_numeric(pmi_df['PMI'], errors='coerce')
    return pmi_df

# Função para plotar os gráficos
def plot_economic_data(cambio, inflacao, pmi):
    plt.figure(figsize=(14, 8))

    # Normalizando os dados para que as escalas sejam comparáveis
    cambio_scaled = cambio['Adj Close'] / cambio['Adj Close'].max() * 100
    pmi_scaled = pmi['PMI'] / pmi['PMI'].max() * 100
    inflacao_scaled = inflacao['FP.CPI.TOTL.ZG'] / inflacao['FP.CPI.TOTL.ZG'].max() * 100

    # Centralizando as linhas do câmbio e PMI
    cambio_centered = cambio_scaled - (cambio_scaled.mean() - inflacao_scaled.mean())
    pmi_centered = pmi_scaled - (pmi_scaled.mean() - inflacao_scaled.mean())

    # Plot da inflação como gráfico de área
    plt.fill_between(inflacao.index, inflacao_scaled, color='red', alpha=0.5, label='Inflação no Japão (%)')

    # Plot das linhas para câmbio e PMI
    plt.plot(cambio.index, cambio_centered, label='Taxa de Câmbio USD/JPY (Centralizada)', color='blue', linewidth=2)
    plt.plot(pmi.index, pmi_centered, label='PMI Industrial Japão (Centralizado)', color='green', linewidth=2)

    # Configurações de título e rótulos
    plt.title('Comparação: Taxa de Câmbio, Inflação e PMI Industrial do Japão', fontsize=16)
    plt.xlabel('Data', fontsize=12)
    plt.ylabel('Valores Normalizados (%)', fontsize=12)
    plt.axhline(0, color='black', linewidth=0.8, linestyle='--')  # Linha horizontal no zero
    plt.ylim(-30, 110)  # Ajuste dos limites do eixo y para visualização melhor
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# 1. Taxa de câmbio USD/JPY (usando yfinance)
usd_jpy = yf.download('JPY=X', start='2015-01-01', end=datetime.today().strftime('%Y-%m-%d'))

# 2. Inflação no Japão (usando a API do Banco Mundial)
inflation_data = get_world_bank_data(indicator='FP.CPI.TOTL.ZG', country='JP')

# 3. PMI Industrial do Japão (lendo do Excel)
pmi_data = get_pmi_from_excel('/content/drive/MyDrive/Files/PMI.xlsx')

# Chamando a função de plotagem com dados reais
plot_economic_data(usd_jpy, inflation_data, pmi_data)
```

```python
import yfinance as yf
import pandas as pd
from datetime import datetime
import requests
from openpyxl import Workbook

# Função para obter dados da API do Banco Mundial
def get_world_bank_data(indicator, country='JP'):
    url = f"https://api.worldbank.org/v2/country/{country}/indicator/{indicator}?format=json&date=2015:2024"
    response = requests.get(url)
    data = response.json()

    if len(data) > 1:
        df = pd.DataFrame(data[1])
        df['date'] = pd.to_datetime(df['date'])
        df.set_index('date', inplace=True)
        df.rename(columns={'value': indicator}, inplace=True)
        return df[[indicator]]
    else:
        print(f"Nenhum dado encontrado para o indicador {indicator}.")
        return pd.DataFrame()

# Função para criar a tabela com os dados e salvar em formato .xlsx
def create_economic_data_table(sony_stock, cambio, inflacao):
    # Renomear colunas para evitar conflito
    sony_stock = sony_stock[['Adj Close']].rename(columns={'Adj Close': 'Preço da Ação Sony (JPY)'})
    cambio = cambio[['Adj Close']].rename(columns={'Adj Close': 'Taxa de Câmbio USD/JPY'})

    # Definindo o índice de datas que vamos usar
    all_dates = pd.date_range(start='2015-01-01', end=datetime.today().strftime('%Y-%m-%d'))

    # Combinando todos os dados em um único DataFrame
    combined_data = pd.DataFrame(index=all_dates)
    combined_data = combined_data.join(sony_stock, how='left').join(cambio, how='left').join(inflacao, how='left')

    # Exibir a tabela
    print("\nTabela com dados econômicos:")
    print(combined_data)

    # Salvar a tabela em um arquivo Excel
    combined_data.to_excel('economic_data_summary.xlsx', engine='openpyxl')
    print("Arquivo 'economic_data_summary.xlsx' foi salvo com sucesso.")

# 1. Preço da Ação da Sony (usando yfinance)
sony_stock = yf.download('6758.T', start='2015-01-01', end=datetime.today().strftime('%Y-%m-%d'))

# 2. Taxa de câmbio USD/JPY (usando yfinance)
usd_jpy = yf.download('JPY=X', start='2015-01-01', end=datetime.today().strftime('%Y-%m-%d'))

# 3. Inflação no Japão (usando a API do Banco Mundial)
inflation_data = get_world_bank_data(indicator='FP.CPI.TOTL.ZG', country='JP')

# Chamando a função para criar a tabela com dados reais
create_economic_data_table(sony_stock, usd_jpy, inflation_data)
```

```python
# Importando as bibliotecas necessárias
import yfinance as yf
import pandas as pd

# Definindo os tickers das empresas
tickers = ['6758.T', '6723.T', '6503.T']

# Criando um DataFrame vazio para armazenar os dados
data = []

# Iterando sobre cada ticker para coletar os dados
for ticker in tickers:
    # Obtendo os dados da empresa
    stock = yf.Ticker(ticker)

    # Coletando informações financeiras
    info = stock.info

    # Extraindo a Market Cap
    market_cap = info.get('marketCap', None)

    # Coletando dados financeiros
    financials = stock.financials

    # Verificando se as informações financeiras estão disponíveis
    if financials is not None:
        revenue = financials.loc['Total Revenue']  # Receita
        ebitda = financials.loc['EBITDA']  # EBITDA

        # Adicionando dados para os anos disponíveis
        for year in revenue.index:
            data.append({
                'Ticker': ticker,
                'Year': year,
                'Market Cap': market_cap,
                'Receita': revenue[year],
                'EBITDA': ebitda.get(year, None)  # Usando get para evitar KeyError
            })

# Criando um DataFrame a partir da lista de dados
df = pd.DataFrame(data)

# Exibindo a tabela final
df
```

```python
import yfinance as yf
import matplotlib.pyplot as plt

# Baixar os dados financeiros da Sony
ticker = "SONY"
sony_data = yf.Ticker(ticker)

# Obter dados financeiros relevantes
cash_flow = sony_data.cashflow
balance_sheet = sony_data.balance_sheet

# Verificar os dados disponíveis no fluxo de caixa
print("Cash Flow Data:")
print(cash_flow)

# Calcular o fluxo de caixa livre (FCF) para 2024
try:
    fcf_2024 = cash_flow.loc['Free Cash Flow'][0]  # Usando 'Free Cash Flow'
except KeyError as e:
    print(f"Erro ao acessar dados de fluxo de caixa: {e}")
    fcf_2024 = 0  # Valor padrão em caso de erro

# Projeções para os anos seguintes
growth_rate = 0.05  # Assumindo um crescimento constante de 5% ao ano
fcf_projection = [fcf_2024 * ((1 + growth_rate) ** i) for i in range(7)]  # De 2024 a 2030

# Calcular a dívida total
try:
    total_debt = balance_sheet.loc['Total Debt'][0]
except KeyError as e:
    print(f"Erro ao acessar dados do balanço patrimonial: {e}")
    total_debt = 0  # Valor padrão em caso de erro

# Projeção de amortização da dívida
amortization_rate = 0.1  # Supondo que 10% do fluxo de caixa livre será usado para amortização
debt_amortization_projection = [fcf * amortization_rate for fcf in fcf_projection]

# Criar gráfico de barras verticais
years = list(range(2024, 2031))
plt.figure(figsize=(10, 6))
bars = plt.bar(years, debt_amortization_projection, color='#ee4954', alpha=0.7)  # Azul puro

# Adicionar valores formatados acima das barras
for bar in bars:
    yval = bar.get_height()
    # Formatação dos valores
    formatted_value = f'{yval / 1e9:.2f}B' if yval >= 1e9 else f'{yval / 1e6:.2f}M' if yval >= 1e6 else f'{int(yval)}'
    plt.text(bar.get_x() + bar.get_width()/2, yval, formatted_value, ha='center', va='bottom')

plt.axhline(0, color='red', linewidth=0.8, linestyle='--')  # Linha zero para referência
plt.xlabel('Ano')
plt.ylabel('Valor da Amortização (USD)')
plt.xticks(years)

# Remover linhas de grade
plt.grid(False)

# Definindo fundo transparente
plt.gca().set_facecolor('none')  # Fundo do gráfico transparente
plt.gcf().patch.set_alpha(0)      # Fundo da figura transparente

# Exibir o gráfico sem título
plt.tight_layout()
plt.show()
```

```python
import yfinance as yf
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression

# Obter dados da Sony
sony = yf.Ticker("6758.T")
financials = sony.financials

# Extrair receitas e custos
receitas = financials.loc['Total Revenue'].values
custo_bens_vendidos = financials.loc['Cost Of Revenue'].values

# Converter para numérico (em caso de strings)
receitas = pd.to_numeric(receitas, errors='coerce')
custo_bens_vendidos = pd.to_numeric(custo_bens_vendidos, errors='coerce')

# Tratar valores ausentes
receitas = np.nan_to_num(receitas)  # Substitui NaN por 0
custo_bens_vendidos = np.nan_to_num(custo_bens_vendidos)  # Substitui NaN por 0

# Garantir que não haja divisão por zero
receitas = np.where(receitas == 0, 1e-9, receitas)

# Calcular margem bruta
margem_bruta = (receitas - custo_bens_vendidos) / receitas * 100

# Verificando se há NaN após o cálculo
if np.isnan(margem_bruta).any():
    raise ValueError("Existem valores NaN na margem bruta.")

# Anos correspondentes
anos = np.arange(2019, 2024)

# Projeção até 2028
modelo_custo = LinearRegression()
modelo_margem = LinearRegression()

# Reshape dos dados para o modelo
anos_reshape = anos.reshape(-1, 1)

# Treinando modelos de regressão
modelo_custo.fit(anos_reshape, custo_bens_vendidos.reshape(-1, 1))
modelo_margem.fit(anos_reshape, margem_bruta.reshape(-1, 1))

# Previsão para os anos futuros
anos_futuros = np.arange(2024, 2029).reshape(-1, 1)
custo_projetado = modelo_custo.predict(anos_futuros)
margem_projetada = modelo_margem.predict(anos_futuros)

# Concatenar dados reais e projetados
custos_totais = np.concatenate([custo_bens_vendidos, custo_projetado.flatten()])
margem_total = np.concatenate([margem_bruta, margem_projetada.flatten()])

# Anos totais
anos_totais = np.concatenate([anos, anos_futuros.flatten()])

# Criar gráfico com fundo transparente
fig, ax1 = plt.subplots(figsize=(12, 6))

# Gráfico de barras para custos com transparência
ax1.bar(anos_totais, custos_totais, color='#FF5733', alpha=0.3, edgecolor='black', label='Custos (bilhões)')
ax1.tick_params(axis='y', labelcolor='#FF5733')

# Criar segundo eixo para margem bruta
ax2 = ax1.twinx()
ax2.plot(anos_totais, margem_total, color='#3357FF', marker='o', alpha=0.7, label='Margem Bruta (%)')
ax2.tick_params(axis='y', labelcolor='#3357FF')

# Linhas de grade
ax1.yaxis.grid(True, linestyle='--', alpha=0.5)
ax2.yaxis.grid(True, linestyle='--', alpha=0.5)

# Títulos e legendas
ax1.set_title('Custos e Margem Bruta da Sony (2019-2028)', fontsize=16)
ax1.set_xticks(anos_totais)
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Remover títulos dos eixos
ax1.set_ylabel('')
ax2.set_ylabel('')

# Definir fundo da figura como transparente
fig.patch.set_alpha(0.0)  # Fundo da figura
ax1.set_facecolor((1, 1, 1, 0))  # Fundo dos eixos
ax2.set_facecolor((1, 1, 1, 0))  # Fundo dos eixos

# Salvar a figura com fundo transparente
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from numpy_financial import npv  # Importando a função npv da biblioteca correta

# Parâmetros
custo_dívida = 0.03  # 3%
custo_capital_proprio = 0.08  # 8%
participacao_divida = 0.4  # 40%
participacao_capital_proprio = 0.6  # 60%

# Cálculo do WACC
wacc = (custo_capital_proprio * participacao_capital_proprio) + (custo_dívida * participacao_divida * (1 - 0.3))  # Considerando uma alíquota de imposto de 30%
print(f"WACC: {wacc:.2%}")

# Valores de WACC e crescimento
wacc_variacao = np.linspace(0.02, 0.1, 5)  # Variação do WACC entre 2% e 10%
crescimento_variacao = np.linspace(0.02, 0.1, 5)  # Variação do crescimento entre 2% e 10%

# Criar matriz de sensibilidade
sensibilidade_matriz = np.zeros((len(wacc_variacao), len(crescimento_variacao)))

# Calcular valor presente para cada combinação de WACC e taxa de crescimento
for i, wacc_value in enumerate(wacc_variacao):
    for j, crescimento in enumerate(crescimento_variacao):
        # Fluxos de caixa com crescimento anual
        fluxo_caixa = np.array([-100] + [30 * (1 + crescimento) ** k for k in range(5)])  # Fluxos de caixa projetados
        valor_presente = npv(wacc_value, fluxo_caixa)  # Cálculo do Valor Presente
        sensibilidade_matriz[i, j] = valor_presente  # Armazenar o valor presente na matriz

# Criar DataFrame para a matriz de sensibilidade
sensibilidade_df = pd.DataFrame(sensibilidade_matriz, columns=[f'Crescimento {c*100:.1f}%' for c in crescimento_variacao],
                                 index=[f'WACC {w*100:.1f}%' for w in wacc_variacao])

# Exibir a matriz de sensibilidade
print(sensibilidade_df)

# Criar gráfico da matriz de sensibilidade
plt.figure(figsize=(10, 6))
plt.imshow(sensibilidade_matriz, cmap='viridis', aspect='auto')
plt.colorbar(label='Valor Presente')
plt.xticks(ticks=np.arange(len(crescimento_variacao)), labels=[f'Crescimento {c*100:.1f}%' for c in crescimento_variacao])
plt.yticks(ticks=np.arange(len(wacc_variacao)), labels=[f'WACC {w*100:.1f}%' for w in wacc_variacao])
plt.title('Matriz de Sensibilidade: WACC vs Valor Presente', fontsize=14)
plt.xlabel('Taxa de Crescimento', fontsize=12)
plt.ylabel('WACC', fontsize=12)
plt.gca().set_facecolor('none')  # Fundo do gráfico transparente
plt.gcf().patch.set_alpha(0)  # Fundo da figura transparente
plt.tight_layout()
plt.show()
```
